```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV data
df = pd.read_csv('../../data/combined_data/gvcomix_data.csv')

# Plot the data
plt.figure(figsize=(10, 6))
plt.scatter(df['oilreserves'], df['milexp_pergdpSIPRI'])
plt.xlabel('Oil Reserves')
plt.ylabel('Military expenditure')
plt.title('Military expenditure vs Oil Reserves')
plt.grid(True)
plt.ylim(0, 50)
plt.show()
```
